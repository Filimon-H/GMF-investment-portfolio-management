```python
# Standard imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os
sys.path.append(os.path.abspath('../src'))

# Our custom module
from data_loader import fetch_asset_data
from preprocessing import inspect_asset, clean_missing_values, normalize_close_column

# Plot settings
plt.style.use("seaborn-v0_8-darkgrid")
sns.set_context("notebook")
```

```python
# Define tickers and dates
tickers = ['TSLA', 'SPY', 'BND']
start_date = '2015-07-01'
end_date = '2025-07-31'

# Fetch data safely
data_dict, adj_close = fetch_asset_data(tickers, start_date, end_date)

# Save to CSV
import os
os.makedirs("../data/raw", exist_ok=True)

for ticker in tickers:
    if ticker in data_dict:
        data_dict[ticker].to_csv(f"../data/raw/{ticker}.csv")

adj_close.to_csv("../data/raw/adj_close.csv")
```
